```python
import numpy as np
import random
import matplotlib.pyplot as plt
from google.colab import files
```

```python
def finance_map(n):

  x = np.zeros(n+1, dtype=float)
  x[0] = 0.1

  alpha = 0.9

  y = np.zeros(n+1, dtype=float)
  y[0] = 0.1

  for i in range(1, n+1):
    y[i] = y[i-1] - alpha*np.tan(x[i-1])
    x[i] = np.sin(x[i-1]) + y[i]

  plt.plot(x, y)
  plt.show()

  return [x,y]
```

```python
finance_map(1000)
```

```python
def modified_logistic(n):
  x = np.zeros(n+1, dtype=float)
  x[0] = 0.1

  a = 1.4
  b = 0.3

  y = np.zeros(n+1, dtype=float)
  y[0] = 0.1

  for i in range(1, n+1):
    y[i] = b**2 * x[i-1]
    x[i] = x[i-1] **2 + y[i-1]**2 - a**2

  plt.plot(x, y)
  plt.show()

  return [x,y]
```

```python
modified_logistic(1000)
```

```python
def eye_map(n,x_0,y_0, alpha):
  x = np.zeros(n, dtype=float)
  x[0] = x_0

  y = np.zeros(n, dtype=float)
  y[0] = y_0

  for i in range(1, n):
    y[i] = np.tan(y[i-1]) - alpha * np.sin(x[i-1])
    x[i] = np.sin(x[i-1]) + np.tan(y[i])

  return [x,y]
```

```python
def galaxy_map(n):
  x = np.zeros(n+1, dtype=float)
  x[0] = 0.1

  alpha = 0.9

  y = np.zeros(n+1, dtype=float)
  y[0] = 0.1

  for i in range(1, n+1):
    y[i] = np.sin(y[i-1]) - alpha * np.tan(x[i-1])
    x[i] = np.tan(x[i-1]) + np.sin(y[i])

  plt.plot(x, y)
  plt.show()

  return [x,y]
```
